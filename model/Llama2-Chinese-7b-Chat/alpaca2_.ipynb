```python
import os

#選擇要使用的gpu 若不設定就是使用全部可用資源
# os.environ["CUDA_VISIBLE_DEVICES"] = "0"
import torch
import torch.nn as nn
import bitsandbytes as bnb
from datasets import load_dataset
import transformers
# auto系列套件可以自動辨認參數指定的模型架構來提供相應的訓練資料
from transformers import AutoTokenizer, AutoConfig, LlamaForCausalLM
from datasets import DatasetDict


# 載入lora和quatizaiton套件
from peft import prepare_model_for_int8_training, LoraConfig, get_peft_model
```

```python
# RTX4090顯卡的訓練超參數設置

# 單個GPU在梯度更新一次計算的樣本數大小，若有提供GRADIENT_ACCUMULATION_STEPS和BATCH_SIZE則可忽略此參數。
MICRO_BATCH_SIZE = 4  # change to 4 for RTX4090
# 一次
BATCH_SIZE = 32
# 在一次梯度更新之前需要累積幾個micr_batch_size的梯度計算資料
GRADIENT_ACCUMULATION_STEPS = BATCH_SIZE // MICRO_BATCH_SIZE
EPOCHS = 5  # paper uses 3
LEARNING_RATE = 3e-4  # from the original paper
CUTOFF_LEN = 1024  # 256 accounts for about 96% of the data
LORA_R = 8
LORA_ALPHA = 16
LORA_DROPOUT = 0.1
VAL_SET_SIZE = 0.2
```

```python
    if VAL_SET_SIZE > 0:
        train_testvalid = data['train'].train_test_split(test_size=VAL_SET_SIZE, seed=42)
        # Split the 10% test + valid in half test, half valid
        test_valid = train_testvalid['test'].train_test_split(test_size=0.5, seed=42)
        # gather everyone if you want to have a single DatasetDict
        data = DatasetDict({
            'train': train_testvalid['train'],
            'test': test_valid['test'],
            'valid': test_valid['train']})

        train_data = train_testvalid["train"].shuffle().map(generate_and_tokenize_prompt)
        val_data = test_valid["test"].shuffle().map(generate_and_tokenize_prompt)
    else:
        train_data = data['train'].shuffle().map(generate_and_tokenize_prompt)
        val_data = None
```

```python
# 每筆資料都會padding到1024個token input_ids labels attention_mask
print(len(train_data[5000]["labels"]))
```

```python
# 載入quatizaiton後的hf模型

model = prepare_model_for_int8_training(model)
```

```python
# lora設定
config = LoraConfig(
    r=LORA_R,
    lora_alpha=LORA_ALPHA,
    #target_modules=["q_proj", "v_proj"],
    lora_dropout=LORA_DROPOUT,
    bias="none",
    task_type="CAUSAL_LM",
)
# 載入lora設定到模型
model = get_peft_model(model, config)
```

```python

# fp16半精度福點數 相較fp32儲存量減少一半
# logging steps 顯示每步的損失
# 每幾步儲存training logs
trainer = transformers.Trainer(
    model=model,
    train_dataset=train_data,
    eval_dataset=val_data,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=MICRO_BATCH_SIZE,
        gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
        warmup_steps=100,
        num_train_epochs=EPOCHS,
        learning_rate=LEARNING_RATE,
        fp16=True,
        logging_steps=50,
        output_dir="/root/checkpoint/", #/root/checkpoint/
        save_total_limit=10,
        save_steps=100,
        report_to="wandb",
        run_name= "llama2_law2",
        eval_steps=100 if VAL_SET_SIZE > 0 else None,
        evaluation_strategy="steps" if VAL_SET_SIZE > 0 else "no",
        save_strategy="steps",
        # remove_unused_columns=False,
        load_best_model_at_end=True if VAL_SET_SIZE > 0 else False
    ),
    # 準備訓練樣本的套件 可以對樣本做padding和對模型做masked label traning
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)
# 模型在回覆問題時是否緩存上一個問題的計算內容
model.config.use_cache = False
# 載入上一個訓練的checkpoint繼續訓練
trainer.train(resume_from_checkpoint= True)


model.save_pretrained("/root/config") #/root/config
```

```python
## 資料集沒有instruction  config.json也沒有eos 所以不用-1

from transformers import set_seed, GenerationConfig
import os
import argparse
from peft import PeftModel, PeftConfig
# os.environ["CUDA_VISIBLE_DEVICES"] = "0"
from colorama import Fore, Style
from transformers import AutoTokenizer, AutoModelForCausalLM

def set_seeds(seed):
    set_seed(seed)
    torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)

def generate_prompt(input):
    return f"""你是人工智慧助理，以下是用戶的問題。你要對用戶的問題提供有用、安全、詳細和禮貌的回答。
### human:
{input}
### gpt:"""

def evaluate(input, generation_config):
    prompt = generate_prompt(input)
    inputs = tokenizer(prompt, return_tensors="pt")
    input_ids = inputs["input_ids"].cuda()
    generation_output = model.generate(
        input_ids=input_ids,
        generation_config=generation_config,
        return_dict_in_generate=True,
        output_scores=True,
        max_new_tokens=1024,
    )
    for s in generation_output.sequences:
        output = tokenizer.decode(s)
        print(f"{Fore.GREEN}回覆:{Style.RESET_ALL}")
        print(output.split("### gpt:")[1].strip() + '\n')

def parse_config():
    parser = argparse.ArgumentParser()
    parser.add_argument("--model_name", type=str, default="yentinglin/Taiwan-LLM-7B-v2.1-chat")
    parser.add_argument("--ckpt_name", type=str, default='stevenaaa1207/Llama2_Taiwan_Law')
    parser.add_argument("--cache_dir", type=str, default="/content/cache")

    parser.add_argument("--seed", type=int, default=42)
    parser.add_argument("--temperature", type=float, default=0)
    parser.add_argument("--top_p", type=float, default=0.9)
    parser.add_argument("--num_beams", type=int, default=1)
    parser.add_argument("--no_repeat_ngram_size", type=int, default=3)

    return parser.parse_args([])

```

```python
args = parse_config()
print(args)
```

```python
if __name__ == '__main__':
    args = parse_config()
    # set_seeds(args.seed)

    tokenizer = AutoTokenizer.from_pretrained("yentinglin/Taiwan-LLM-7B-v2.1-chat")
    # load from checkpoint
    model = PeftModel.from_pretrained(model, args.ckpt_name, device_map={'': 0}, load_in_8bit=True)

    generation_config = GenerationConfig(
        temperature=args.temperature,
        top_p=args.top_p,
        num_beams=args.num_beams,
        no_repeat_ngram_size=args.no_repeat_ngram_size,
    )

    while(True):
        evaluate(input(f"\n{'-'*10}\n{Fore.BLUE}指令: {Style.RESET_ALL}"), generation_config)
```
